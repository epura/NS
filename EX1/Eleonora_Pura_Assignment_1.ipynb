```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

<h1>A01.1 Building and visualising networks</h1>

```python
nodes_nr = 32
```

<h2>Star Network</h2>

```python
g1 = nx.Graph()
g1.add_node(nodes_nr-1)

for i in range(nodes_nr-1):
    g1.add_node(i)
    g1.add_edge(nodes_nr-1, i)
```

```python
#Check
g1.number_of_nodes()
```

```python
g1.nodes()
```

```python
nx.draw(g1, with_labels = True)
```

<h2>Fully-connected</h2>

```python
g2 = nx.Graph()
for i in range(nodes_nr):
    g2.add_node(i)
    for j in range(g2.number_of_nodes()):
        if i != j:
            g2.add_edge(i, j)
```

```python
#Check
g2.number_of_nodes()
```

```python
g2.degree()
```

```python
nx.draw(g2, with_labels=True)
```

<h2>One-dimensional lattice with periodic boundary conditions where each node  has degree  = 2</h2>


```python
g3 = nx.Graph()

for i in range(nodes_nr-1):
    g3.add_node(i)
    g3.add_edge(i,i+1)

g3.add_edge(nodes_nr-1, 0)
```

```python
g3.number_of_nodes()
```

```python
g3.degree()
```

```python
nx.draw(g3, with_labels=True)
```

<h1>A01.2 Global Properties</h1>

```python
#Functions

def max_degree(network):
    max = ()
    for degree in network.degree():
        if max==(): max = degree
        if degree[1] > max[1]: max = degree
    return max

def show_degree_distribution(degrees):
    plt.title("Degree Distributions")
    plt.xlabel("k")
    plt.ylabel("P(K)")
    my_bins = np.logspace(np.log10(min(degrees)),np.log10(max(degrees) + 1), 30) 
    plt.hist(degrees, bins = my_bins, density=True, color='green')
    plt.xscale('log')
    plt.yscale('log')
    plt.style.use('classic')
    plt.show()

    
```

<h3>Zachary</h3>

```python
g_zachary = nx.read_gml("data/graph_karate.gml")
```

```python
degrees = [degree[1] for degree in g_zachary.degree()]
avg_degree = np.average(degrees)
density = avg_degree/(g_zachary.number_of_nodes()-1)
```

```python
print("Average degree: {}".format(round(avg_degree, 3)))
print("Density: {}".format(round(density, 4)))
print("Max degree (id, degree): {}".format(max_degree(g_zachary)))
```

```python
show_degree_distribution(degrees)
```
